```python
import numpy as np
import math
import matplotlib.pyplot as plt

def sigmoid(x):
    a = []
    for item in x:
        a.append(1.0/(1.0 + math.exp(-item)))
    return a
x = np.arange(-10, 10, 0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

xx=np.arange(-10, 10, 1)
yy=[math.exp(-x) for x in xx]
yy_=[math.exp(x) for x in xx]
print(type(yy))
print(xx,yy)

plt.plot(xx,yy,'b-')
plt.plot(xx,yy_,'r-')
plt.show()
```

```python
'''
鸢尾分类
'''
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

#导入数据
from sklearn import datasets

iris = datasets.load_iris()

#只取一个维度
x_1=iris['data'][:,:1];
print(x_1)
print(iris['target'])
y_1=(iris['target']==1).astype(int) #iris['target']返回数组，然后对数组中的每个元素与1进行对比，返回True或False，astype函数是将布尔值 True/False 转成 1/0
print(y_1)
print(type(iris['target']))

lr=LogisticRegression(max_iter=10000)
lr.fit(x_1, y_1)
result = lr.predict_proba(np.array([[5.9]]))
print(result)
for i in range(result.shape[0]):
    if result[i][0]>result[i][1]:
        print('predict 5.9,result:0')
    else:
        print('predict 5.9,result:1')


'''
使用二分类实现多分类的思路：
    每次以每个分类作为正例，其余分类作为负例，求出其最优模型
    当有一个需要预测的样本时，将其带入每个模型中其为正例的概率，概率最大的分类即为当前预测样本的
    预测结果
'''
```

```python
'''
softmax算法：
    - 多类别分类算法
    - 多项式分布，多项概率之和为1
'''
import numpy as np
from sklearn.linear_model import LogisticRegression

#导入数据
from sklearn import datasets

iris = datasets.load_iris()

#只取一个维度
x_1=iris['data'];
print(x_1)
y_1=iris['target']
print(y_1)
#softmax多分类算法
softmax=LogisticRegression(max_iter=10000,multi_class='multinomial')
softmax.fit(x_1, y_1)
result = softmax.predict_proba(np.array([[5.9,3.,  5.1 ,1.8]]))

x_y=np.c_[x_1,y_1]
print(x_y)
print(result)




```
