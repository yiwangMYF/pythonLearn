```python
#饼图
man=71351
woman=68187
man_perc=man/(woman+man)
woman_perc=woman/(woman+man)
#添加名称
labels=['男','女']
#添加颜色
colors=['blue','red']
#绘制饼状图 pie
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
# labels 名称 colors：颜色，explode=分裂 autopct 显示百分比
paches,texts,autotexts=plt.pie([man_perc,woman_perc],labels=labels,colors=colors,explode=(0,0.05),autopct='%0.1f%%')
#设置饼状图中的字体颜色
for text in autotexts:
    text.set_color('white')
#设置字体大小
for text in texts+autotexts:
    text.set_fontsize(20)
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# 生成数据
data = np.random.normal(loc=0, scale=1, size=1000)

# 使用seaborn绘制正态分布的密度图
sns.kdeplot(data, shade=True, color='blue')

# 添加标题
plt.title("Normal Distribution (Kernel Density Estimation)")

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
mean = 0     # 均值
std_dev = 1  # 标准差

# 生成 x 值，通常在均值附近选取区间
x = np.linspace(-5, 5, 1000)

# 计算正态分布的概率密度函数（PDF）
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(- 0.5 * ((x - mean) / std_dev) ** 2)

# 绘制图形
plt.plot(x, pdf, label='Normal Distribution', color='blue')

# 添加标题和标签
plt.title("Normal Distribution (Mean=0, Std=1)")
plt.xlabel("X")
plt.ylabel("Density")

# 显示网格
plt.grid(True)

# 显示图例
plt.legend()

# 显示图形
plt.show()
```
